```python
import libs.style_discrimination as st
import libs.beliveability_discrimination as be
import libs.meaning_discrimination as me
```

```python
%%time
be_model = be.load('./models/')
```

```python
%%time
st_model = st.load('./models/')
```

```python
%%time
tokenizer, morph, w2v = me.load('./models/')
```

```python
import json
with open('./data/seed_sample.json', encoding='utf-8') as f:
    data = json.load(f)
```

```python
seed = list(data.keys())[0]
generated = list(data.values())[0]
```

```python
%%time
be_probs = be.beliveability(generated, be_model)
```

```python
%%time
st_probs = st.checkovness(generated, st_model)
```

```python
%%time
me_probs = me.sentence_similarity(seed, generated, tokenizer, morph, w2v)
```

```python
topn = 20
print('seed_phrase:\n\n', seed, '\n' + '_'*20 + '\n')
print('topn believable sentences:\n\n', '\n'.join(map(str, be.top_best(generated, be_probs, topn=topn))), '\n' + '_'*20 + '\n')
print('topn checkov sentenses:\n\n', '\n'.join(map(str, be.top_best(generated, st_probs, topn=topn))), '\n' + '_'*20 + '\n')
print('topn meaning sentences:\n\n', '\n'.join(map(str, be.top_best(generated, me_probs, topn=topn))), '\n' + '_'*20 + '\n')
```
